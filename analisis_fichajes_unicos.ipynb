```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# Leer archivo .txt
# Reemplaza 'archivo.txt' con la ruta a tu archivo
ruta_archivo = 'fichadas.txt'

columns = ['lugar','fecha','hora','tarjeta','legajo','nombre','estado','x', 'y']

df = pd.read_csv(ruta_archivo, sep=',',encoding='latin1', header=None, names=columns)


df = df.drop(['x','y'], axis=1)

df['fecha'] = pd.to_datetime(df['fecha'], format='%Y/%m/%d')

# Convertir la columna 'hora' a tipo datetime.time (solo hora)
df['hora'] = pd.to_datetime(df['hora'], format='%H:%M:%S').dt.time


display(df)

df.columns
```

```python
df_entrada = df[df['estado'] == 40]
df_salida = df[df['estado'] == 41]


df_entrada = df_entrada.sort_values(by=['fecha', 'hora'])
df_salida = df_salida.sort_values(by=['fecha', 'hora'])


df_entrada_unico = df_entrada.drop_duplicates(subset=['legajo', 'fecha'])
df_salida_unico = df_salida.drop_duplicates(subset=['legajo', 'fecha'])

display(df_entrada_unico)
display(df_salida_unico)

# df_unico = df.drop_duplicates(subset=['legajo', 'fecha', 'estado'])

# df_unico.head(40)
```

```python
entrada_resumen = df_entrada_unico.groupby(['legajo', 'fecha']).size().reset_index(name='entradas')
salida_resumen = df_salida_unico.groupby(['legajo', 'fecha']).size().reset_index(name='salidas')

# Fusionar los dos DataFrames en uno solo
df_resumen = pd.merge(entrada_resumen, salida_resumen, on=['legajo', 'fecha'], how='outer').fillna(0)

# Ahora, puedes agregar una nueva columna para identificar si el legajo hizo solo una entrada, solo una salida, o ambas
df_resumen['estado'] = df_resumen.apply(
    lambda row: 'Solo entrada' if row['entradas'] > 0 and row['salidas'] == 0 else (
                'Solo salida' if row['entradas'] == 0 and row['salidas'] > 0 else 'Entrada y salida'), axis=1
)

# Mostrar el resumen
display(df_resumen)


df_resumen.to_csv('un_solo_fichaje.txt', sep=',', index=False)
```
